8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.

```python
X_tmp = X.copy()
X_tmp.loc[X_tmp['EthnicGroup'].isna(),'EthnicGroup']=X_tmp.loc[:,'EthnicGroup'].mode()[0]
X_tmp.loc[X_tmp['ParentMaritalStatus'].isna(),'ParentMaritalStatus']=X_tmp.loc[:,'ParentMaritalStatus'].mode()[0]
X_tmp.loc[X_tmp['WklyStudyHours'].isna(),'WklyStudyHours']=X_tmp.loc[:,'WklyStudyHours'].mode()[0]
X_tmp.isna().sum()
```

```python
X_tmp.loc[X_tmp.ParentEduc.isna(),'ParentEduc']='no formal education'
X_tmp.isna().sum()
```

```python
X_tmp.loc[X_tmp.TestPrep.isna(),'TestPrep']='none'
X_tmp.isna().sum()
```

```python
X_cleaned=X_tmp.copy()
X_cleaned=X_cleaned.loc[X_cleaned['PracticeSport'].notnull() & X_cleaned['IsFirstChild'].notnull(), :]
X_cleaned.isna().sum()
```

```python
X_cleaned.loc[X_cleaned['NrSiblings'].isna(),'NrSiblings']=X_cleaned.loc[:,'NrSiblings'].median()
X_cleaned.isna().sum()
```

```python
X_cleaned.loc[X_cleaned.TransportMeans.isna(),'TransportMeans']='walking'
X_cleaned.isna().sum()
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
print(X_cleaned['Gender'].value_counts())
```

```python
X_cleaned['Gender'] = X_cleaned['Gender'].astype(str)
X_cleaned['LunchType'] = X_cleaned['LunchType'].astype(str)
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first', dtype=int)
encoded_data = encoder.fit_transform(X_cleaned[['Gender', 'LunchType']])
X_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Gender', 'LunchType']))

X_cleaned.reset_index(drop=True, inplace=True) 
X_encoded.reset_index(drop=True, inplace=True)
X_clean = pd.concat([X_cleaned, X_encoded], axis=1)
X_clean
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_ethnic = LabelEncoder()
label_encoder_parent_educ = LabelEncoder()
label_encoder_test_prep = LabelEncoder()
label_encoder_parent_marital = LabelEncoder()
label_encoder_practice_sport = LabelEncoder()
label_encoder_is_first_child = LabelEncoder()
label_encoder_transport_means = LabelEncoder() 
label_encoder_wkly_study_hours = LabelEncoder()


X_clean['EthnicGroup'] = label_encoder_ethnic.fit_transform(X_clean['EthnicGroup'])
X_clean['ParentEduc'] = label_encoder_parent_educ.fit_transform(X_clean['ParentEduc'])
X_clean['TestPrep'] = label_encoder_test_prep.fit_transform(X_clean['TestPrep'])
X_clean['ParentMaritalStatus'] = label_encoder_parent_marital.fit_transform(X_clean['ParentMaritalStatus'])
X_clean['PracticeSport'] = label_encoder_practice_sport.fit_transform(X_clean['PracticeSport'])
X_clean['IsFirstChild'] = label_encoder_is_first_child.fit_transform(X_clean['IsFirstChild'])
X_clean['TransportMeans'] = label_encoder_transport_means.fit_transform(X_clean['TransportMeans'])
X_clean['WklyStudyHours'] = label_encoder_wkly_study_hours.fit_transform(X_clean['WklyStudyHours'])

X_clean
```
